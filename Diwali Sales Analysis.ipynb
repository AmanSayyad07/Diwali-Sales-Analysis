```python
## EDA On Gender Column
```

```python
ax=sns.countplot(x='Gender',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ax=sns.barplot(x='Gender',y='Amount',data=sales_gen)

```

From the above graph we can see that the most of the buyers are the females and purchasing power of the female is greater than men

```python
##  EDA  on Age Group
```

```python
ax=sns.countplot(x='Age Group',data=df,hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#total Amount vs Age Group
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ax=sns.barplot(x='Age Group',y='Amount',data=sales_gen)
```

```python
#  Total Orders By States
sales_gen=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
ax=sns.barplot(x='State',y='Orders',data=sales_gen)
sns.set(rc={'figure.figsize':(18,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#  Total Amount By States
sales_gen=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
ax=sns.barplot(x='State',y='Amount',data=sales_gen)
sns.set(rc={'figure.figsize':(15,5)})
```

 From above graph we can see that the UP, Maharashtra, Karanataka are the most contributer for our sales

 Analyzing the geopolitical conditions we have to increase our sales in other states

 From first state graph we can know that the Kerla has the orders but it has not given the sales , We have to also take a look on that 

```python
ax=sns.countplot(x='Occupation',data=df)

sns.set(rc={'figure.figsize':(18,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Occupation',y='Amount',data=sales_gen)
sns.set(rc={'figure.figsize':(10,5)})
```

 From above graph we can see that the IT, Healthcare,Aviation are the highest sales genarator occupations 
 

```python
ax=sns.countplot(x='Product_Category',data=df)

sns.set(rc={'figure.figsize':(27,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(x='Product_Category',y='Amount',data=sales_gen)
sns.set(rc={'figure.figsize':(20,6)})
```

 From the first Product Category graph we can see that the Clothing is  the highest in terms of Orders But 
 Not in terms of sales
